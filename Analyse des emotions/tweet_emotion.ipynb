```
#importation du jeu de données

tweet = spark.sql("select * from training_1600000_processed_noemoticon_efba6_csv")
```

```
display(tweet.describe())
```

```
from pyspark.sql import functions as F
```

```
#Détermination de la répartition des smiley - le datset est balanced

rep = tweet.groupBy(tweet.sentiment).agg(F.count(tweet.sentiment).alias("Repartition"))
```

```
rep.show()
```

```
t_rdd = tweet.rdd.map(tuple)
for i in t_rdd.take(10): print(i)
```

```
# Détermination de la distribution de la longueur des Tweets

trdd = t_rdd.map(lambda o: (int(o[0]),o[1],o[2],o[3],o[4],o[5],int(len((o[5]).split(" ")))))
tweet2 = trdd.toDF(schema=["sentiment","_c1","_c2","_c3","_c4","tweet","nbe_mot"])
```

```
tweet2.show(25)
```

```
# Feature engineering : vectorisation des tweet. Dans chaque vecteur, une dimension correspond à un mot. Le nombre d'occurence de chaque mot apparaît.

from pyspark.ml.feature import HashingTF, IDF, Tokenizer
from pyspark.ml.feature import CountVectorizer
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.linalg import Vectors
from pyspark.ml import Pipeline

tokenizer = Tokenizer(inputCol="tweet", outputCol="words")
wordsData = tokenizer.transform(tweet2)

hashingTF = HashingTF(inputCol="words", outputCol="features")
hashingTF_model = hashingTF.transform(wordsData)
print("Out of hashingTF function")
hashingTF_model.select('words',col('features').alias('Features(vocab_size,[index],[tf])')).show(truncate=False)

lr = LogisticRegression(maxIter=30, regParam=0.001,labelCol="sentiment")
pipeline = Pipeline(stages=[tokenizer, hashingTF, lr])

```

```
# Segmentation de l'échantillon en deux jeux de données train (70%) et test (30%)

training, test = hashingTF_model.randomSplit([0.7, 0.3])

# Calcul des paramètres du modèle de régression logistique
lrModel = lr.fit(training)
```

```
# Réalisation des prédictions

predictions = lrModel.transform(test)
selected = prediction.select("sentiment", "tweet", "prediction")
for row in selected.collect():
    print(row)
```

```
#Evaluation du modèle : calcul du taux d'erreur de la regression logistique.

from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Select (prediction, true label) and compute test error
evaluator = MulticlassClassificationEvaluator(
    labelCol="sentiment", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test Error = %g " % (1.0 - accuracy))
```
